```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Assuming df is your DataFrame

# Ensure 'Place' is numeric
df['Place'] = pd.to_numeric(df['Place'], errors='coerce')

# Drop rows with missing values in 'Place', 'TotalKg', or 'Wilks'
df = df.dropna(subset=['Place', 'TotalKg', 'Wilks'])

# Calculate correlation coefficients
correlation_place_totalkg = pearsonr(df['Place'], df['TotalKg'])
correlation_place_wilks = pearsonr(df['Place'], df['Wilks'])

# Print correlation coefficients
print(f'Correlation between Place and TotalKg: {correlation_place_totalkg}')
print(f'Correlation between Place and Wilks: {correlation_place_wilks}')


plt.figure(figsize=(14,9))
plt.subplot(1,2,1)
sns.scatterplot(x='Place',y='TotalKg',data=df)
plt.title('TotalKg by Place')
          
          
plt.subplot(1,2,2)
sns.scatterplot(x='Place',y='Wilks',data=df)
plt.title('Wilks by Place')
          
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame and it has 'Place', 'Division', 'BestSquatKg', 'BestBenchKg', and 'BestDeadliftKg' columns

# Ensure 'Place' is numeric
df['Place'] = pd.to_numeric(df['Place'], errors='coerce')

# Drop rows with missing 'Place' values
df = df.dropna(subset=['Place'])

# Find top 10 competitors in each division
top_10_competitors = df.groupby('Division').apply(lambda x: x.nsmallest(10, 'Place')).reset_index(drop=True)

# Calculate mean performance metrics for top 10 competitors in each division
grouped_by_division = top_10_competitors.groupby('Division')[['BestSquatKg', 'BestBenchKg', 'BestDeadliftKg']].mean().reset_index()

# Melt the DataFrame for easier plotting
melted_df = grouped_by_division.melt(id_vars='Division', var_name='Lift', value_name='Weight')

# Plot the results
plt.figure(figsize=(12, 8))
sns.barplot(x='Division', y='Weight', hue='Lift', data=melted_df, palette='viridis')
plt.title('Mean Performance of Top 10 Competitors by Division')
plt.xlabel('Division')
plt.ylabel('Weight (Kg)')
plt.legend(title='Lift')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
